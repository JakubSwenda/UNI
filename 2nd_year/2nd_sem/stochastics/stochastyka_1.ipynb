```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

## sposob pierwszy

```python
lista_punktow=[]

for i in range (0,100):
    x=random.randrange(0,11)
    lista_punktow.append(x)
plt.plot(lista_punktow)
plt.show()
srednia=np.mean(lista_punktow)
print("srednia to ok.: ",srednia)
```

## sposob drugi

```python
lista=[]
for i in range (0,11):
    lista.append(i)

lista_pkt=[]
for i in range (0,100):
    x=random.choice(lista)
    lista_pkt.append(x)
    
plt.plot(lista_pkt)
plt.show()
srednia=np.mean(lista_pkt)
print("srednia to ok.: ",srednia)
```

## sposob trzeci

```python
punkty=[]
for i in range (0,100):
    x=random.uniform(0,1)
    punkty.append(x)
plt.plot(punkty)
plt.show()
srednia=np.mean(punkty)
print("srednia to ok.: ",srednia)
```

# 1D

```python
punkty=[]
for i in range (0,100):
    x=random.uniform(0,1)
    punkty.append(x)
    
x2=[]

x=0

for i in punkty:
    if i < 0.5:
        x=x-1
        x2.append(x)
    elif i > 0.5:
        x=x+1
        x2.append(x)
plt.plot(x2)
```

# 2D

```python
punkty=[]
for i in range (0,6000):
    x=random.uniform(0,1)
    punkty.append(x)
    
x2=[]
y2=[]

x=0
y=0

for i in punkty:
    if i < 0.25:
        x=x+1
        y=y+1
        x2.append(x)
        y2.append(y)
    elif i >= 0.25 and i < 0.5:
        x=x+1
        y=y-1
        x2.append(x)
        y2.append(y)
    elif i >= 0.5 and i < 0.75:
        x=x-1
        y=y-1
        x2.append(x)
        y2.append(y)
    elif i >= 0.75 and i < 1:
        x=x-1
        y=y+1
        x2.append(x)
        y2.append(y)
plt.plot(x2,y2)
plt.show()
```
