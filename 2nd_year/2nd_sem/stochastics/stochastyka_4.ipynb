# model Monte Carlo

```python
import random as rd
import matplotlib.pyplot as plt
```

```python
if wybor == "parzysta" or "PARZYSTA" or "Parzysta":
    def wybor_liczby():
        liczba=rd.randint(1,100)
        if liczba%2!=0 or liczba==10:
            return False
        elif liczba%2==0:
            return True
elif wybor=="nieparzysta" or "NIEPARZYSTA" or "Nieparzysta":
    def wybor_liczby():
        liczba=rd.randint(1,100)
        if liczba%2==0 or liczba==11:
            return False
        elif liczba%2==1:
            return True
    
```

```python
def gra(suma_pieniedzy,zaklad,ile_gier):
    ilosc_gier=[]
    pieniadze=[]
    gra=1
    for gra in range(ile_gier):
        if wybor_liczby():
            suma_pieniedzy=suma_pieniedzy+(zaklad*1.02)
            ilosc_gier.append(gra)
            pieniadze.append(suma_pieniedzy)
        else:
            suma_pieniedzy=suma_pieniedzy-zaklad
            ilosc_gier.append(gra)
            pieniadze.append(suma_pieniedzy)

    plt.ylabel('Pieniadze gracza')
    plt.xlabel('Ilosc zakladow')
    plt.plot(ilosc_gier,pieniadze)
    
    ostateczne_fundusze.append(pieniadze[-1])
    return ostateczne_fundusze
```

```python
wybor=input("parzysta czy nieparzysta?\n")
x=int(input("Ile pieniedzy chcesz przeznaczyc na gre?[$]\n"))
y=int(input("Ile pieniedzy chcesz przeznaczyc na jeden zaklad?[$]\n"))
z=int(input("Ile zakladow chcesz rozegrac?\n"))
```

```python
ostateczne_fundusze=[]

for i in range (100):
    koncowe_fundusze=gra(x,y,z)
print(koncowe_fundusze)
print(sum(koncowe_fundusze))

print(f"Gracz rozpoczal z {x} $")
print("Gracz na koniec mia≈Ç",str(sum(koncowe_fundusze)/len(koncowe_fundusze)),"$")
```
