```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
#znalezc prstwo znalezienia sie w stanach zacztynajac ze stanow po 15 krokach
macierz=np.array([[0,0.5,0.5,0,0],
                  [0.5,0,0,0.5,0],
                  [0,0,0.7,0.3,0],
                  [0,0,0.5,0.5,0],
                  [1,0,0,0,0]])
n=9 #kroki
p0=[1,0,0,0,0]
p1=[0,0,1,0,0]
p2=[0,0,0,0,1]
macierz
```

```python
pkroki=[]
def kroki(start,transition,steps):
    transition=np.ndarray(shape=(5,5), dtype=float)
    for i in range(5):
        for j in range(5):
            macierz[i][j]=transition[i][j]
            pkroki.append(macierz[i][j])
    return start@np.linalg.matrix_power(transition,steps)
```

```python
q1=kroki(p0,macierz,n)
q2=kroki(p1,macierz,n)
q3=kroki(p2,macierz,n)
print(q1)
print(q2)
print(q3)
```

##### stan sk jest osiagalny ze stanu sj jesli pjk(n)>0 dla pewnego n(max9)
##### stany sk i sj nazywamy wzajemnie komunikujacymi sie jesli stan sk jest osiagalny ze stany sj i odwrotnie

```python
Mrx = np.array([[0, 0.5, 0.5, 0, 0], 
                [0.5, 0, 0, 0.5, 0], 
                [0, 0, 0.7, 0.3, 0], 
                [0, 0, 0.5, 0.5, 0], 
                [1, 0, 0, 0, 0]])

p0 = [1, 0, 0, 0, 0]

num_steps = 9

prob0_list = []
prob1_list = []
prob2_list = []
prob3_list = []
prob4_list = []

for i in range(num_steps):
    prob = np.linalg.matrix_power(Mrx, i)
    prob0 = np.dot(p0, prob)
    prob1 = prob[1][2]
    prob2 = prob[4][0]
    prob3 = prob[2][3]
    prob4 = prob[3][3]
    prob0_list.append(prob0)
    prob1_list.append(prob1)
    prob2_list.append(prob2)
    prob3_list.append(prob3)
    prob4_list.append(prob4)



plt.plot(range(num_steps), prob1_list, label='Stan 1')
plt.plot(range(num_steps), prob2_list, label='Stan 2')
plt.plot(range(num_steps), prob3_list, label='Stan 3')
plt.plot(range(num_steps), prob4_list, label='Stan 4')
plt.xlabel('Kroki')
plt.ylabel('Prawdopodobienstwo')
plt.title('Prawdopodobienstwo bycia w danym stanie w czasie')
plt.legend()
plt.show()
```
