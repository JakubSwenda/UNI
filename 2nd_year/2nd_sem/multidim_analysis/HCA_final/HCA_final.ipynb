```python
listacord = []
lista_cord_name = []
macierz = iter(range(len(Y), 200))
full_di = {}
Z = Y.copy()
lista_min = []

while Y.size >= 4:
    triangle = np.tril(Y)
    triangle_flat = np.array(triangle).flatten()
    min1 = min(i for i in triangle_flat if i>0)
    coord = np.where(triangle == min1)
    listacord.append(list(coord))
    coord1 = max(coord)[0]
    coord2 = min(coord)[0]
    listacord2 = Y.iloc[coord1].name, Y.iloc[coord2].name
    lista_cord_name.append(listacord2)
    lista_cord_name
    lista_min.append(min1)

    
    new_name = next(macierz)
    
    nazwy_col = []
    
    
    for i in range(0, len(Z)):
        nazwy_col.append(Z.columns[i])

    if Y.iloc[coord1].name in nazwy_col and Y.iloc[coord2].name in nazwy_col:
        full_di[new_name] = 2
        
    elif (Y.iloc[coord1].name in nazwy_col and Y.iloc[coord2].name not in nazwy_col):
        full_di[new_name] = 1 + full_di.get(Y.iloc[coord2].name)
        
    elif (Y.iloc[coord1].name not in nazwy_col and Y.iloc[coord2].name in nazwy_col):
        full_di[new_name] = 1 + full_di.get(Y.iloc[coord1].name)
        
    else:
        full_di[new_name] = full_di.get(Y.iloc[coord1].name) + full_di.get(Y.iloc[coord2].name)
        
        
    
    wektor = []
    
    
    for i in range(len(Y)):
        if Y.iloc[coord1, i] == 0:
            wektor.append(Y.iloc[coord1, i])
        
        elif Y.iloc[coord2, i] == 0:
            wektor.append(Y.iloc[coord2, i])
        
        elif Y.iloc[coord2, i] > Y.iloc[coord1, i]:
            wektor.append(Y.iloc[coord2, i])
        
        else:
            wektor.append(Y.iloc[coord1, i])
                    
    Y.iloc[coord2] = wektor
    Y.iloc[:, coord2] = wektor
    Y.drop(Y.columns[coord1], axis=1, inplace=True)
    Y.drop(Y.index[coord1], axis=0, inplace=True)
    Y.rename(columns={Y.columns[coord2]:new_name},inplace=True)
    Y.rename(index={Y.index[coord2]:new_name},inplace=True)
```
