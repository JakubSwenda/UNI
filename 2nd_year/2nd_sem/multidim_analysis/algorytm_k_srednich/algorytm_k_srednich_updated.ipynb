```python
wspolrzedne={lista_dystansow_punkt1[0]:lista_punktow_PC2[0],
             lista_dystansow_punkt2[1]:lista_punktow_PC2[1],
             lista_dystansow_punkt3[2]:lista_punktow_PC2[2]}
```

```python
p1 = np.array(lista_dystansow_punkt1)
p2 = np.array(lista_dystansow_punkt2)
p3 = np.array(lista_dystansow_punkt3)

p1 = p1.reshape(30,1)
p2 = p2.reshape(30,1)
p3 = p3.reshape(30,1)
```

```python
joined = np.concatenate((p1,p2,p3), axis = 1)
```

```python
joined
```

```python
column_indices = []
min_values = []

for i in range(len(joined)):
    row = joined[i]
    min_value = np.min(row)
    min_values.append(min_value)
    column_index = np.argmin(row)
    column_indices.append(column_index)
```

```python
lis1 = []
lis2 = []
lis3 = []

for i in range(0, 30):
    if column_indices[i] == 0:
        lis1.append(min_values[i])
    elif column_indices[i] == 1:
        lis2.append(min_values[i])
    elif column_indices[i] == 2:
        lis3.append(min_values[i])
```
