```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
dane=pd.read_excel('macierz_do_k_srednich.xlsx',index_col=0)
```

```python
dane
```

```python
lista=['PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10']
```

```python
dane=dane.drop(lista,axis=1)
```

```python
min1=min(dane['PC1'])
max1=max(dane['PC1'])
min2=min(dane['PC2'])
max2=max(dane['PC2'])
```

```python
lista_punktow_PC1=[]
for i in range(0,3):
    x=rd.uniform(min1, max1)
    lista_punktow_PC1.append(x)
```

```python
lista_punktow_PC1
```

```python
lista_punktow_PC2=[]
for i in range(0,3):
    x=rd.uniform(min2, max2)
    lista_punktow_PC2.append(x)
```

```python
lista_punktow_PC2
```

```python
wspolrzedne={lista_punktow_PC1[0]:lista_punktow_PC2[0],
             lista_punktow_PC1[1]:lista_punktow_PC2[1],
             lista_punktow_PC1[2]:lista_punktow_PC2[2]}
```

```python
wspolrzedne
```

```python
result = wspolrzedne.items()
data = list(result)
wspolrzedne_array = np.array(data)
print(wspolrzedne_array)
```

```python
print(dane)
```

```python
lista_dystansow_punkt1=[]
lista_dystansow_punkt2=[]
lista_dystansow_punkt3=[]
```

```python
for i in range(0,30):
    point1=np.array((dane['PC1'][i],dane['PC2'][i]))
    point2=wspolrzedne_array[0]
    distance=np.linalg.norm(point1 - point2)
    lista_dystansow_punkt1.append(distance)
for i in range(0,30):
    point1=np.array((dane['PC1'][i],dane['PC2'][i]))
    point2=wspolrzedne_array[1]
    distance=np.linalg.norm(point1 - point2)
    lista_dystansow_punkt2.append(distance)
for i in range(0,30):
    point1=np.array((dane['PC1'][i],dane['PC2'][i]))
    point2=wspolrzedne_array[2]
    distance=np.linalg.norm(point1 - point2)
    lista_dystansow_punkt3.append(distance)
```

```python
print(lista_dystansow_punkt1,'\n',
lista_dystansow_punkt2,'\n',
lista_dystansow_punkt3,'\n')
```
