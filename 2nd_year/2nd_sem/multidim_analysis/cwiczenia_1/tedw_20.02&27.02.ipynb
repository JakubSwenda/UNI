# 1

```python
A=[]
i=0
while i<10:
    x=random.randrange(1,20)
    A.append(x)
    i=i+1
wektorA=np.vstack(A)
print(wektorA)
```

# 2

```python
B=[]
i=0
while i<10:
    x=random.uniform(0,1)
    B.append(x)
    i=i+1
wektorB=np.vstack(B)
print(wektorB)
```

# 3

```python
wektorC=wektorA*wektorB
print(wektorC)
```

```python
c=2
wektor4=wektorA*c
print(wektor4)
```
