```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
def autoskalowanie(xi,xm,xstd,lista):
    Zxi=(xi-xm)/xstd
    lista.append(Zxi)
    return lista
```

```python
dane=pd.read_excel("dane_bio.xlsx",index_col=0)

x1m=st.mean(dane["X1"])
x1std=st.stdev(dane["X1"])
x2m=st.mean(dane["X2"])
x2std=st.stdev(dane["X2"])
x3m=st.mean(dane["X3"])
x3std=st.stdev(dane["X3"])
x4m=st.mean(dane["X4"])
x4std=st.stdev(dane["X4"])
x5m=st.mean(dane["X5"])
x5std=st.stdev(dane["X5"])
x6m=st.mean(dane["X6"])
x6std=st.stdev(dane["X6"])
x7m=st.mean(dane["X7"])
x7std=st.stdev(dane["X7"])
x8m=st.mean(dane["X8"])
x8std=st.stdev(dane["X8"])
x9m=st.mean(dane["X9"])
x9std=st.stdev(dane["X9"])
x10m=st.mean(dane["X10"])
x10std=st.stdev(dane["X10"])
X1auto=[]
X2auto=[]
X3auto=[]
X4auto=[]
X5auto=[]
X6auto=[]
X7auto=[]
X8auto=[]
X9auto=[]
X10auto=[]

for i in dane['X1']:
    y=autoskalowanie(i,x1m,x1std,X1auto)
for i in dane['X2']:
    y=autoskalowanie(i,x2m,x2std,X2auto)
for i in dane['X3']:
    y=autoskalowanie(i,x3m,x3std,X3auto)
for i in dane['X4']:
    y=autoskalowanie(i,x4m,x4std,X4auto)
for i in dane['X5']:
    y=autoskalowanie(i,x5m,x5std,X5auto)
for i in dane['X6']:
    y=autoskalowanie(i,x6m,x6std,X6auto)
for i in dane['X7']:
    y=autoskalowanie(i,x7m,x7std,X7auto)
for i in dane['X8']:
    y=autoskalowanie(i,x8m,x8std,X8auto)
for i in dane['X9']:
    y=autoskalowanie(i,x9m,x9std,X9auto)
for i in dane['X10']:
    y=autoskalowanie(i,x10m,x10std,X10auto)

    
matrixauto=pd.DataFrame(X1auto,columns=["ZX1"])

matrixauto['ZX2'] = X2auto
matrixauto['ZX3'] = X3auto
matrixauto['ZX4'] = X4auto
matrixauto['ZX5'] = X5auto
matrixauto['ZX6'] = X6auto
matrixauto['ZX7'] = X7auto
matrixauto['ZX8'] = X8auto
matrixauto['ZX9'] = X9auto
matrixauto['ZX10'] = X10auto

odleglosci=np.zeros((30,30))
df_odl=pd.DataFrame(odleglosci)
df_odl1=pd.DataFrame(odleglosci)
matrixauto.index=['A1','A2','A3','A4','A5','B1','B2','B3','B4','B5','C1','C2','C3','C4','C5',
                  'D1','D2','D3','D4','D5','E1','E2','E3','E4','E5','F1','F2','F3','F4','F5']
```

```python
matrixauto
```

```python
def cov_1(x,y):
    x_mean=x.mean()
    y_mean=y.mean()
    cov_id=sum((x_i-x_mean)*(y_i-y_mean) for x_i,y_i in zip(x,y))/np.sqrt(sum((x_i-x_mean)**2 for x_i in x)*
                                                                  sum((y_i-y_mean)**2 for y_i in y))
    return cov_id
```

```python
n=len(dane.columns)
cov_2=np.zeros(shape=(n,n))
```

```python
for i in range(n):
    for j in range(n):
        cov_2[i,j]=cov_1(dane.iloc[:,i], dane.iloc[:,j])
cov_3=pd.DataFrame(cov_2)
```

```python
cov_3
```

```python
eigen_vals,eigen_vecs=np.linalg.eig(cov_3)
```

```python
print('wartosci\n',eigen_vals,'\n\nwektory\n',eigen_vecs)
```
