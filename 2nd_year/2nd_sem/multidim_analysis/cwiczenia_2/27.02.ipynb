```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dane=pd.DataFrame(np.random.normal(0,1,24))
dane
```

```python
dane2=pd.DataFrame(np.random.normal(20,5,1))
dane2
```

```python
dane3=dane.append(dane2, ignore_index=True)
dane3
```

```python
dane_bez=dane3.drop(24)
dane_bez
```

```python
m=dane_bez.mean()
s=np.std(dane_bez)
print("srednia=",m,"\nodchylenie=",s)
```

```python
from scipy.stats import t
```

```python
alpha=0.05
df=25
ttest=t.ppf(1-alpha/2,df-1)
ttest
```

```python
xmin=m-ttest*s
xmax=m+ttest*s
xmin,xmax
#punkt 17(ten odstajacy) nie miesci sie w przedziale ufnosci wiec nalezy go usunac
```

# Cwiczenie 2

```python
matrix1=np.random.randint(10,size=(3,3))
matrix2=np.random.randint(10,size=(3,3))

print("macierz 1\n",matrix1,"\n","\nmacierz 2\n",matrix2)

dodawanie=matrix1+matrix2
print("\nmacierze dodane\n",dodawanie)

odejmowanie=matrix1-matrix2
print("\nmacierze odjete\n",odejmowanie)

mnozenie=np.dot(matrix1,matrix2)
print("\nmacierze mnozone\n",mnozenie)

sumakol_matrix1=np.sum(matrix1,axis=0)
print("\nmacierz 1, suma kolumn\n",sumakol_matrix1)

sumakol_matrix2=np.sum(matrix2,axis=0)
print("\nmacierz 2, suma kolumn\n",sumakol_matrix2)

sumawiersz_matrix1=np.sum(matrix1,axis=1)
print("\nmacierz 1, suma wierszy\n",sumawiersz_matrix1)

sumawiersz_matrix2=np.sum(matrix2,axis=1)
print("\nmacierz 2, suma wierszy\n",sumawiersz_matrix2)

print("\nmacierz 1, inwersja\n",np.linalg.inv(matrix1))

print("\nmacierz 2, inwersja\n",np.linalg.inv(matrix2))
```
