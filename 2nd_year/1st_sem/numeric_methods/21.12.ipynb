```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
```

```python
fs=100  #Hz
dt=1/fs
czas=np.arange(0,1,dt)
f=200
f2=350
sin1=np.sin(2*pi*f*czas)
sin2=np.sin(2*pi*f2*czas)
```

```python
plt.plot(czas,sin1+sin2)
plt.show()
```

```python
#transformata fouriera
fourier=np.fft.fft(sin1+sin2)

#normalizacja
fouriernorm=fourier/len(sin1+sin2)

#plt.plot(fouriernorm)
#plt.show

#generowanie osi x
freq=np.fft.fftfreq(len(sin1+sin2))*fs

#bierzemy modul z tranformaty
plt.plot(freq,abs(fouriernorm))
plt.show
```
