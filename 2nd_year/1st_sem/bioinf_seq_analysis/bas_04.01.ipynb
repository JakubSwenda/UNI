```python
import requests as rq
```

```python
def kegginfo(database):
    url=f'https://rest.kegg.jp/info/{database}'
    resp=rq.get(url)
    print(resp.text)
```

```python
def kegglist(database, org=''):
    url=f'https://rest.kegg.jp/list/{database}/{org}'
    resp=rq.get(url)
    return resp.text
```

```python
for record in kegglist('pathway','hsa').split('\n'):
    if 'fatty acid' in record.lower():
        print(record)
```

```python
def keggfind(database, query):
    url=f'https://rest.kegg.jp/find/{database}/{query}'
    resp=rq.get(url)
    return resp.text
```

```python
for genome in (keggfind('genome','eco')).split('\n'):
    if 'escherichia coli' in genome.lower():
        print(genome)
```

```python
def keggget(dbentries, option=''):
    url=f'https://rest.kegg.jp/get/{dbentries}/{option}'
    resp=rq.get(url)
    return resp.text
```

```python
print(keggfind('compound','cocaine'))
```

```python
print(keggfind('drug','cocaine'))
```

```python
print(keggget('dr:D00110'))
```

```python
print(keggget('cpd:C01416'))
```
